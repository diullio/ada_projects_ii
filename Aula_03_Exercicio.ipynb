```python
import numpy as np
import pandas as pd
import requests
```

```python
def requisicao(pais):
    request = requests.get(f'http://universities.hipolabs.com/search?country={pais}')
    requisicao = request.json()
    df = pd.DataFrame(requisicao)
    return df
```

```python
def dataframe():
    df2 = pd.concat([requisicao("Brazil"), requisicao("Portugal")], join="outer")
    return df2
```
