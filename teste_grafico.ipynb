```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("EDA_Industrializados_202110.csv", sep=";", encoding="latin1")
```

```python
filtro = df.loc[(df['UF_VENDA'] == "TO") & (df['MUNICIPIO_VENDA'] == "TAGUATINGA")]
```

```python
filtro
```

```python
soma = filtro.groupby("PRINCIPIO_ATIVO").sum()
```

```python
soma
```

```python
y = filtro[filtro["QTD_VENDIDA"] > 10]
```

```python
y
```

```python
x = y["PRINCIPIO_ATIVO"]	
```

```python
z = y["QTD_VENDIDA"]
```

```python
x
```

```python
z
```

```python
barras = plt.bar(x, z, color='blue')
plt.title("Medicametos vendidos em Taguatinga - TO")
plt.xlabel("Medicamentos")
plt.ylabel("Quantidade")
plt.xticks(rotation=30, ha='right')
```
